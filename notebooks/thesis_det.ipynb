```python
# SARSA Delays
sarsa_0_avg, sarsa_0_std = load.stats_train(method='sarsa', source='Results', test_type='Delay0')
sarsa_3_avg, sarsa_3_std = load.stats_train(method='sarsa', source='Results', test_type='Delay3')
sarsa_5_avg, sarsa_5_std = load.stats_train(method='sarsa', source='Results', test_type='Delay5')
sarsa_10_avg, sarsa_10_std = load.stats_train(method='sarsa', source='Results', test_type='Delay10')
sarsa_15_avg, sarsa_15_std = load.stats_train(method='sarsa', source='Results', test_type='Delay15')
sarsa_20_avg, sarsa_20_std = load.stats_train(method='sarsa', source='Results', test_type='Delay20')

#DSARSA Delays
dsarsa_3_avg, dsarsa_3_std = load.stats_train(method='dsarsa', source='Results', test_type='Delay3')
dsarsa_5_avg, dsarsa_5_std = load.stats_train(method='dsarsa', source='Results', test_type='Delay5')
dsarsa_10_avg, dsarsa_10_std = load.stats_train(method='dsarsa', source='Results', test_type='Delay10')
dsarsa_15_avg, dsarsa_15_std = load.stats_train(method='dsarsa', source='Results', test_type='Delay15')
dsarsa_20_avg, dsarsa_20_std = load.stats_train(method='dsarsa', source='Results', test_type='Delay20')

# TRPO Delays
trpo_0_avg, trpo_0_std = load.stats_train(method='trpo', source='Augmented', test_type='Delay0', epoch=2000)
trpo_3_avg, trpo_3_std = load.stats_train(method='trpo', source='Augmented', test_type='Delay3', epoch=2000)
trpo_5_avg, trpo_5_std = load.stats_train(method='trpo', source='Augmented', test_type='Delay5', epoch=2000)
trpo_10_avg, trpo_10_std = load.stats_train(method='trpo', source='Augmented', test_type='Delay10', epoch=2000)
trpo_15_avg, trpo_15_std = load.stats_train(method='trpo', source='Augmented', test_type='Delay15', epoch=2000)
trpo_20_avg, trpo_20_std = load.stats_train(method='trpo', source='Augmented', test_type='Delay20', epoch=2000)

# M-TRPO Delays
mtrpo_3_avg, mtrpo_3_std = load.stats_train(method='trpo', source='Memoryless', test_type='Delay3', epoch=2000)
mtrpo_5_avg, mtrpo_5_std = load.stats_train(method='trpo', source='Memoryless', test_type='Delay5', epoch=2000)
mtrpo_10_avg, mtrpo_10_std = load.stats_train(method='trpo', source='Memoryless', test_type='Delay10', epoch=2000)
mtrpo_15_avg, mtrpo_15_std = load.stats_train(method='trpo', source='Memoryless', test_type='Delay15', epoch=2000)
mtrpo_20_avg, mtrpo_20_std = load.stats_train(method='trpo', source='Memoryless', test_type='Delay20', epoch=2000)

# L2-TRPO Delays
l2trpo_3_avg, l2trpo_3_std = load.stats_train(method='l2trpo', source='Results-BIG', test_type='Delay3-BIG', epoch=2000)
l2trpo_5_avg, l2trpo_5_std = load.stats_train(method='l2trpo', source='Results-BIG', test_type='Delay5-BIG', epoch=2000)
l2trpo_10_avg, l2trpo_10_std = load.stats_train(method='l2trpo', source='Results-BIG', test_type='Delay10-BIG', epoch=2000)
l2trpo_15_avg, l2trpo_15_std = load.stats_train(method='l2trpo', source='Results-BIG', test_type='Delay15-BIG', epoch=2000)
l2trpo_20_avg, l2trpo_20_std = load.stats_train(method='l2trpo', source='Results-BIG', test_type='Delay20-BIG', epoch=2000)

# D-TRPO Delays
dtrpo_3_avg, dtrpo_3_std = load.stats_train(method='dtrpo', source='Results', test_type='Delay3', epoch=2000)
dtrpo_5_avg, dtrpo_5_std = load.stats_train(method='dtrpo', source='Results', test_type='Delay5', epoch=2000)
dtrpo_10_avg, dtrpo_10_std = load.stats_train(method='dtrpo', source='Results', test_type='Delay10', epoch=2000)
dtrpo_15_avg, dtrpo_15_std = load.stats_train(method='dtrpo', source='Results', test_type='Delay15', epoch=2000)
```

### Comparison Delay 3

```python
# Plot Range Settings
episode_num = 2000
epochs = np.arange(-1, episode_num, 100)
epochs[0] = 0

# Create Plot Figure
fig = plt.figure(figsize=(8, 5))
ax = fig.add_subplot(1, 1, 1)

# Plot Data
ax.errorbar(epochs, trpo_0_avg[epochs], yerr=trpo_0_std[epochs], color='black', ls='dotted',
            alpha=0.8, linewidth=1.5, ecolor='grey', elinewidth=1, capsize=1, label='Undelayed TRPO')
ax.errorbar(epochs, sarsa_3_avg[epochs], yerr=sarsa_3_std[epochs], color='sienna', ls='dotted',
            alpha=0.8, linewidth=1.5, ecolor='sandybrown', elinewidth=1, capsize=1, label='SARSA')
ax.errorbar(epochs, dsarsa_3_avg[epochs], yerr=dsarsa_3_std[epochs], color='darkorange', ls='dotted',
            alpha=0.8, linewidth=1.5, ecolor='wheat', elinewidth=1, capsize=1, label='D-SARSA')
ax.errorbar(epochs, trpo_3_avg[epochs], yerr=trpo_3_std[epochs], color='purple', ls='dashed',
            alpha=0.9, linewidth=1.5, ecolor='violet', elinewidth=1, capsize=1, label='A-TRPO')
ax.errorbar(epochs, mtrpo_3_avg[epochs], yerr=mtrpo_3_std[epochs], color='forestgreen', ls='dashed',
            alpha=0.9, linewidth=1.5, ecolor='lightgreen', elinewidth=1, capsize=1, label='M-TRPO')
ax.errorbar(epochs, l2trpo_3_avg[epochs], yerr=l2trpo_3_std[epochs], color='royalblue', ls='dashdot',
            alpha=0.9, linewidth=1.5, ecolor='lightblue', elinewidth=1, capsize=1, label='L2-TRPO')
ax.errorbar(epochs, dtrpo_3_avg[epochs], yerr=dtrpo_3_std[epochs], color='red', ls='solid',
            alpha=0.8, linewidth=1.5, ecolor='salmon', elinewidth=1, capsize=1, label='D-TRPO')

# Plot Decoration Settings
ax.legend(fontsize='medium')
ax.set_ylim(-2000, 0)
ax.set_xticks(np.arange(0, 2001, 250))
ax.set_xlabel('Epochs')
ax.set_ylabel('Average Reward')
fig.savefig('./thesis_plots/delay3_comparisons_1.png', dpi=300)
```

## Comparison Stochastic Delays

```python
# Stochastic Delay Process Samples
from utils import delays
import torch
import numpy as np

# D-TRPO Delays
dtrpo_07_avg, dtrpo_07_std = load.stats_train(method='dtrpo', source='AISTATS', test_type='STC-DelayP07')
dtrpo_06_avg, dtrpo_06_std = load.stats_train(method='dtrpo', source='AISTATS', test_type='STC-DelayP06')
dtrpo_05_avg, dtrpo_05_std = load.stats_train(method='dtrpo', source='AISTATS', test_type='STC-DelayP055')
```

### Comparison Delay P=0.7

```python
# Plot Settings
plt.rc('font', family='serif')
plt.rc('xtick', labelsize='x-large')
plt.rc('ytick', labelsize='x-large')
plt.rc('axes', titlesize='x-large')
plt.rc('axes', labelsize='x-large')

# Stochastic Delay Sampled
jump_07 = delays.NormedPositiveCompoundBernoulliProcess(p=0.7, init=0, max_value=50)
jump_07_samples = []
n_samples = 500
for _ in range(n_samples):
    jump_07_samples.append(jump_07.sample()[0])
jump_07_samples = torch.as_tensor(jump_07_samples, dtype=torch.float32)
jump_07_mean = torch.repeat_interleave(torch.mean(jump_07_samples), repeats=n_samples)

# Plot Range Settings
episode_num = 500
epochs = np.arange(-1, episode_num, 10)
epochs[0] = 0

# Create Plot Figure
fig = plt.figure(figsize=(8, 5))
ax = fig.add_subplot(1, 1, 1)

# Plot Data
ax.errorbar(epochs, jump_07_mean[epochs], color='steelblue', ls='dashed',
            alpha=0.8, linewidth=1.0, ecolor='lightblue', elinewidth=1, capsize=1)
ax.errorbar(epochs, jump_07_samples[epochs], color='steelblue', ls='solid',
            alpha=0.8, linewidth=1.5, ecolor='lightblue', elinewidth=1, capsize=1)

# Plot Decoration Settings
ax.legend(fontsize='medium')
ax.set_ylim(0,10)
ax.set_yticks(np.arange(0, 21, 2))
ax.set_xticks(np.arange(0, 501, 50))
ax.set_xlabel('Time-Step')
ax.set_ylabel('Sampled Delay')
fig.savefig('./thesis_plots/delayp07_sampledelay_1.png', dpi=300)
```
