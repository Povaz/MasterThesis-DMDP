Library loading:

```python
from matplotlib import pyplot as plt
from notebooks import loaders as load
import numpy as np
```

Short explanation for method naming scheme:
- SARSA: Standard SARSA implementation learning a memoryless policy in the Delayed MDP
- DSARSA: Delay SARSA implementation learning a memoryless policy in the Delayed MDP
- TRPO: Standard TRPO implementation learning a policy in the the Augmented MDP
- M-TRPO: Standard TRPO implementation learning a memoryless policy in the Delayed MDP
- L2-TRPO: Delay TRPO with Deterministic Belief Module (L2-Loss)
- D-TRPO: Delay TRPO with Stochastic Belief Module (MAF)

## Creating the single Lines for each Algorithm


```python
dsarsa_avg = [dsarsa_3_avg, dsarsa_5_avg, dsarsa_10_avg, dsarsa_15_avg, dsarsa_20_avg]
dsarsa_std = [dsarsa_3_std, dsarsa_5_std, dsarsa_10_std, dsarsa_15_std, dsarsa_20_std]

sarsa_avg = [sarsa_3_avg, sarsa_5_avg, sarsa_10_avg, sarsa_15_avg, sarsa_20_avg]
sarsa_std = [sarsa_3_std, sarsa_5_std, sarsa_10_std, sarsa_15_std, sarsa_20_std]

trpo_avg = [trpo_3_avg, trpo_5_avg, trpo_10_avg, trpo_15_avg, trpo_20_avg]
trpo_std = [trpo_3_std, trpo_5_std, trpo_10_std, trpo_15_std, trpo_20_std]

mtrpo_avg = [mtrpo_3_avg, mtrpo_5_avg, mtrpo_10_avg, mtrpo_15_avg, mtrpo_20_avg]
mtrpo_std = [mtrpo_3_std, mtrpo_5_std, mtrpo_10_std, mtrpo_15_std, mtrpo_20_std]

l2trpo_avg = [l2trpo_3_avg, l2trpo_5_avg, l2trpo_10_avg, l2trpo_15_avg, l2trpo_20_avg]
l2trpo_std = [l2trpo_3_std, l2trpo_5_std, l2trpo_10_std, l2trpo_15_std, l2trpo_20_std]

dtrpo_avg = [dtrpo_3_avg, dtrpo_5_avg, dtrpo_10_avg, dtrpo_15_avg]
dtrpo_std = [dtrpo_3_std, dtrpo_5_std, dtrpo_10_std, dtrpo_15_std]
```

```python
# Plot Range Settings
delays = np.array([3, 5, 10, 15, 20])
jitter = 0.15

# Create Plot Figure
fig = plt.figure(figsize=(8, 5))
ax = fig.add_subplot(1, 1, 1)

# Plot Data
ax.errorbar(delays - jitter*1, sarsa_avg, yerr=sarsa_std, color='darkorange', ls='dotted',
            alpha=0.8, linewidth=2, ecolor='orange', elinewidth=2, capsize=0, label='SARSA(0.9)')
ax.errorbar(delays + jitter*1, dsarsa_avg, yerr=dsarsa_std, color='darkgreen', ls='dotted',
            alpha=0.8, linewidth=2, ecolor='lightgreen', elinewidth=2, capsize=0, label='DSARSA(0.9)')
ax.errorbar(delays, trpo_avg, yerr=trpo_std, color='steelblue', ls='dashdot',
            alpha=0.9, linewidth=2, ecolor='lightblue', elinewidth=2, capsize=0, label='TRPO')
ax.errorbar(delays + jitter*1, mtrpo_avg, yerr=mtrpo_std, color='purple', ls='dashed',
            alpha=0.9, linewidth=2, ecolor='violet', elinewidth=2, capsize=0, label='M-TRPO')
ax.errorbar(delays - jitter*1, l2trpo_avg, yerr=l2trpo_std, color='darkolivegreen', ls='solid',
            alpha=0.9, linewidth=2, ecolor='olive', elinewidth=2, capsize=0, label='L2-TRPO')
ax.errorbar(delays[0:4], dtrpo_avg, yerr=dtrpo_std, color='red', ls='solid',
            alpha=0.9, linewidth=2, ecolor='salmon', elinewidth=2, capsize=0, label='D-TRPO')

# Plot Decoration Settings
ax.legend(fontsize='medium')
ax.set_xticks(delays)
ax.set_yticks(np.arange(-2000, 1, 250))
ax.set_title('Comparison on Test')
ax.set_xlabel('Simulated Delay-Steps')
ax.set_ylabel('Average Reward')
fig.savefig('./plots/Tests_ALL.png', dpi=1000)
```
