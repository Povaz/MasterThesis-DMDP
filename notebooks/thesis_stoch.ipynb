# Section 6.4: Stochastic Environment

## Comparison Noises

```python
from matplotlib import pyplot as plt
from notebooks import loaders as load
import numpy as np
import torch

# Plot Settings
plt.rc('font', family='serif')
plt.rc('xtick', labelsize='large')
plt.rc('ytick', labelsize='large')
plt.rc('axes', titlesize='x-large')
plt.rc('axes', labelsize='x-large')

# Deterministic
l2trpo_5_avg, l2trpo_5_std = load.stats_train(method='l2trpo', source='Results-BIG', test_type='Delay5-BIG', epoch=1000)
dtrpo_5_avg, dtrpo_5_std = load.stats_train(method='dtrpo', source='Results', test_type='Delay5', epoch=1000)

# Beta Noise
dtrpo_beta_avg, dtrpo_beta_std = load.stats_train(method='dtrpo', source='Noises', test_type='Delay5-Beta', epoch=1000)
l2trpo_beta_avg, l2trpo_beta_std = load.stats_train(method='l2trpo', source='Noises', test_type='Delay5-Beta', epoch=1000)

# LogNormal Noise
dtrpo_logn_avg, dtrpo_logn_std = load.stats_train(method='dtrpo', source='Noises', test_type='Delay5-LogNormal', epoch=1000)
l2trpo_logn_avg, l2trpo_logn_std = load.stats_train(method='l2trpo', source='Noises', test_type='Delay5-LogNormal', epoch=1000)

# LogNormalSimple Noise
dtrpo_logns_avg, dtrpo_logns_std = load.stats_train(method='dtrpo', source='Noises', test_type='Delay5-SimpleLogNormal', epoch=1000)
l2trpo_logns_avg, l2trpo_logns_std = load.stats_train(method='l2trpo', source='Noises', test_type='Delay5-LogNormalSimple', epoch=1000)

# Quadratic Noise
dtrpo_quad_avg, dtrpo_quad_std = load.stats_train(method='dtrpo', source='Noises', test_type='Delay5-Quadratic', epoch=1000)
l2trpo_quad_avg, l2trpo_quad_std = load.stats_train(method='l2trpo', source='Noises', test_type='Delay5-Quadratic', epoch=1000)

# Triangular Noise
dtrpo_tri_avg, dtrpo_tri_std = load.stats_train(method='dtrpo', source='Noises', test_type='Delay5-Triangular', epoch=1000)
l2trpo_tri_avg, l2trpo_tri_std = load.stats_train(method='l2trpo', source='Noises', test_type='Delay5-Triangular', epoch=1000)

# Uniform Noise
dtrpo_uni_avg, dtrpo_uni_std = load.stats_train(method='dtrpo', source='Noises', test_type='Delay5-Uniform', epoch=1000)
l2trpo_uni_avg, l2trpo_uni_std = load.stats_train(method='l2trpo', source='Noises', test_type='Delay5-Uniform', epoch=1000)

# UShaped Noise
dtrpo_ush_avg, dtrpo_ush_std = load.stats_train(method='dtrpo', source='Noises', test_type='Delay5-UShaped', epoch=1000)
l2trpo_ush_avg, l2trpo_ush_std = load.stats_train(method='l2trpo', source='Noises', test_type='Delay5-UShaped', epoch=1000)
```

### Beta Noise (8,2)

```python
# Plot Range Settings
episode_num = 1000
epochs = np.arange(-1, episode_num, 50)
epochs[0] = 0

# Create Plot Figure
fig = plt.figure(figsize=(8, 5))
ax = fig.add_subplot(1, 1, 1)

# Plot Data
ax.errorbar(epochs, l2trpo_5_avg[epochs], yerr=l2trpo_5_std[epochs], color='royalblue', ls='dashdot',
            alpha=0.5, linewidth=1.5, ecolor='lightblue', elinewidth=1, capsize=1, label='L2-TRPO Det.')
ax.errorbar(epochs, dtrpo_5_avg[epochs], yerr=dtrpo_5_std[epochs], color='red', ls='solid',
            alpha=0.5, linewidth=1.5, ecolor='salmon', elinewidth=1, capsize=1, label='D-TRPO Det.')
ax.errorbar(epochs, l2trpo_beta_avg[epochs], yerr=l2trpo_beta_std[epochs], color='royalblue', ls='dashdot',
            alpha=0.9, linewidth=1.5, ecolor='lightblue', elinewidth=1, capsize=1, label='L2-TRPO Beta(8,2)')
ax.errorbar(epochs, dtrpo_beta_avg[epochs], yerr=dtrpo_beta_std[epochs], color='red', ls='solid',
            alpha=0.9, linewidth=1.5, ecolor='salmon', elinewidth=1, capsize=1, label='D-TRPO Beta(8,2)')

# Plot Decoration Settings
ax.legend(fontsize='medium')
ax.set_ylim(-2000, 0)
ax.set_xticks(np.arange(0, 1001, 125))
ax.set_xlabel('Epochs')
ax.set_ylabel('Average Reward')
fig.savefig('./thesis_plots/noises_beta_comparison_1.png', dpi=300)
```
